# Data Science Workflow

#### 1.Frame Your Business Problem
    -Is it a regression or a classification problem?
    -Do you have labeled data?
    -NLP?
#### 2.Acquire/Query Data
    -SQL
    -Python (Web Scraping, APIS)
    -Third Party Apps (Power BI, Excel, etc)
    -Other Programming Languages
#### 3a.Clean Data / Feature Engineering
    -Handle Missing Data
    -Parse Data (into Multiple Columns, extract piece of text)
    -Trim Whitespace
    -Convert/Format Datatypes
    -Etc
#### 3b.Exploratory Analysis
    -Check distributions of fields
    -Examine Counts/Check Balance of Data
    -Look at mean, median, mode and other statistics of columns (check for outliers)
    -Look at scatter plots between variables
    -Look at correlations (heatmap) check for multicollinearity
#### 4.Split Data into Train/Test or Crossvalidation
    -Training Data (Data you will train your algorithm on)
    -Test Data (Data you will Evaluate your Model against)
#### 5.Scale your Data (if Necessary)
    -Some columns can have larger ranges than others
    -Scaling allows us to ensure that magnitude doesn't impact our interpretation of the model
    -Scaling can lead to better models
#### 6.Train a Few Algorithms
    -Based on your business problem, pick a few algorithms for regression or classification
    -Train them on your Data
#### 7.Evaluate your models
    -Evaluation methods are different for classification and regression problems
    -Which model performed the best?
    -Evaluate against the test data (unless you used cross validation)
#### 8.Use Best Model to Make Predictions on Unlabled/New Data!!!
#### 9.Backtest and Continually Evaluate How Well Your Model does on Fresh Data

```python
def repl_empty(string):
    if(string == ''):
        return '0'
    else:
        return string
```

```python
# Convert latitude and longitude to floats

pd.options.mode.chained_assignment = None 
covid = covid.loc[(covid.latitude != '') & (covid.longitude != '')]
covid['latitude'] = covid['latitude'].astype(float)
covid['longitude'] = covid['longitude'].astype(float)
covid['confirmed'] = covid['confirmed'].astype(float)
covid['deaths'] = covid['deaths'].apply(repl_empty)
covid['deaths'] = covid['deaths'].astype(float)
covid.dtypes
```

```python
# Write a function to get the season of the year. We expect this would play a role in Covid Deaths

def getUSseasonEquivalent(yearmonth):
    month = int(str(yearmonth)[4:])
    if(month in [12,1,2]):
        season = 'Winter'
    elif(month in [3,4,5]):
        season = 'Spring'
    elif(month in [6,7,8]):
        season = 'Summer'
    elif(month in [9,10,11]):
        season = 'Fall'
    return season 
covid_grouped['season'] = covid_grouped.year_month.apply(getUSseasonEquivalent)
covid_grouped
```

```python
# Do we have all the seasons?

covid_grouped.season.unique()
```

```python
# Prepare to get time series (Previous Month Data)

listmonths = list(covid_grouped.year_month.unique())
listmonths.sort()
listmonths
```

```python
# Function to get previous month deaths

def getpreviousmonthdeaths(yearmonth, country):
    newindex = listmonths.index(yearmonth) - 1
    newyearmonth = listmonths[newindex]
    try:
        prevdeaths = covid_grouped.loc[(covid_grouped.year_month == newyearmonth) & (covid_grouped.country_region == country)].reset_index(drop = True).loc[0,'deaths']
    except:
        prevdeaths = 0
    return prevdeaths

prevdeaths = []
for index, row in covid_grouped.iterrows():
    yearmonth = row['year_month']
    country = row['country_region']
    prevdeaths.append(getpreviousmonthdeaths(yearmonth, country))

covid_grouped['prev_month_deaths'] = prevdeaths
covid_grouped
```

```python
# Function to get Previous Month Confirmed Cases

def getpreviousmonthconfirmed(yearmonth, country):
    newindex = listmonths.index(yearmonth) - 1
    newyearmonth = listmonths[newindex]
    try:
        prevcon = covid_grouped.loc[(covid_grouped.year_month == newyearmonth) & (covid_grouped.country_region == country)].reset_index(drop = True).loc[0,'confirmed']
    except:
        prevcon = 0
    return prevcon

prevcons = []
for index, row in covid_grouped.iterrows():
    yearmonth = row['year_month']
    country = row['country_region']
    prevcons.append(getpreviousmonthconfirmed(yearmonth, country))

covid_grouped['prev_month_confirmed'] = prevcons
covid_grouped
```

# 4.Split Data into Train/Test or Crossvalidation

```python
xdf = cleaned_data.drop('deaths', axis = 1)
ydf = cleaned_data[['deaths']]
X = xdf.values
y = ydf.values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state =52)
```

# 5.Scale your Data (if Necessary)

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scal = scaler.fit_transform(X_train)
X_test_scal = scaler.transform(X_test)
```

```python
# Useful to have standard scaler for support vector machine algorithms

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X_train_svr = sc_X.fit_transform(X_train)
y_train_svr = sc_y.fit_transform(y_train)
```

# 6.Train a Few Algorithms

```python
from sklearn.linear_model import LinearRegression
import numpy as np
regressor = LinearRegression()
regressor.fit(X_train_scal, y_train)

y_pred = regressor.predict(X_test_scal)
np.set_printoptions(precision=2)
#print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))

from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error, mean_squared_error
print('R2: {}'.format(round(r2_score(y_test, y_pred),2)))
print('MAE: {}'.format(round(mean_absolute_error(y_test, y_pred), 2)))
print('RMSE: {}'.format(round(np.sqrt(mean_squared_error(y_test, y_pred)),2)))
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X_train_scal, y_train.ravel())

y_pred = regressor.predict(X_test_scal)
np.set_printoptions(precision=2)
#print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))

print('R2: {}'.format(round(r2_score(y_test, y_pred),2)))
print('MAE: {}'.format(round(mean_absolute_error(y_test, y_pred), 2)))
print('RMSE: {}'.format(round(np.sqrt(mean_squared_error(y_test, y_pred)),2)))
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X_train_scal, y_train)

y_pred = regressor.predict(X_test_scal)
np.set_printoptions(precision=2)
#print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))

print('R2: {}'.format(round(r2_score(y_test, y_pred),2)))
print('MAE: {}'.format(round(mean_absolute_error(y_test, y_pred), 2)))
print('RMSE: {}'.format(round(np.sqrt(mean_squared_error(y_test, y_pred)),2)))
```

# 7.Evaluate your models

XGBoost had the best combination of r2, MAE, and RMSE! It is the winner!

R2 - The explained variance of your model. In other words, with XGboost, our model (with inputs) explains 98% of the variance in covid deaths! A perfect model would explain 100%. So 2% of the variance is likely explained by random chance or variables not in our model. 

MAE - Mean Average Error - This is how much your prediction is off on average. The xgboost was off on average by ~140,000 deaths. This may seem like a lot but if you compare it to the standard deviation of deaths and the sheer volume of deaths for some of these countries, it really isn't bad. 

RMSE - Root Mean Square Error - This is similar to the MAE, except it punishes larger differences more. It takes into account the magnitude of the error. This is important because you could have a model that is rarely off by much, but when it is off, its off by a lot!

# 9.Backtest and Continually Evaluate How Well Your Model does on Fresh Data

Our job is not done! Once we deploy this model in the real world, we need to monitor the results every month to see if it still fits the data well as time passes on! We can do this with the same metrics we used to evaluate our model by simply recording next months deaths for all countries and comparing them to our predictions that we put into place now.
